```python
import requests
import time

def get_user_list():
    res = requests.get("http://localhost:8000/trading/userlist/")
    return json.loads(res.text)['user']

user_list = get_user_list()
print(user_list)


def check_mode():
    trading_user_list = []
    for user in user_list:
        res = requests.get("http://localhost:8000/trading/requirements/{}".format(user))
        result = json.loads(res.text)
        if result['is_trading']:
            trading_user_list.append(result)
            
    return trading_user_list



while True:
    trading_user_list = check_mode()
    
    if len(trading_user_list) > 0:
        print(trading_user_list)
        time.sleep(3)
    
    
    # 트레이딩 알고리즘
    
    
    
    
    
    
    
    
    
    
    
#     if result := check_mode()['is_trading']:
# #         is_trade.append(result['username'])

#         print(result)
    
    
    
# print
        
check_mode()
```
